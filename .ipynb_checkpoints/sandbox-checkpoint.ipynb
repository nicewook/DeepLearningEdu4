```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle = '--', label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin and cos')
plt.legend()
plt.show()
```

```python
def AND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([.5, .5])
    b = -.7
    
    tmp = np.sum(w*x) + b
    
    if tmp > 0:
        return 1
    else:
        return 0
```

```python
print(AND(0,0), AND(0,1), AND(1,0), AND(1,1))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x>0, dtype=np.int)

x = np.arange(-5.0, 5., .1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1 + exp(-x))
```

```python
x = np.array([-1., 1., 2.])
sigmoid(x)
```

```python
x= np.arange(-5., 5., 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-.1, 1.1)
plt.show()
```
